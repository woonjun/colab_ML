```python
import pandas as pd
data2 = pd.read_csv("C:/Users/82107/Documents/GitHub/train_맞춤법검사.csv")
display(data2)
```

```python
import matplotlib.pyplot as plt
grouped_data = data2.groupby('범례')

# 히스토그램을 그리기 위한 서브플롯 생성
fig, axs = plt.subplots(1,len(grouped_data), figsize=(20, 5))

# 각 범례별로 히스토그램 그리기
for (ax, (label, group)) in zip(axs, grouped_data):
    ax.hist(group['result'], bins=range(1, 21), edgecolor='black')
    ax.set_title(f'label: {label}')
    ax.set_xlabel('Result')
    ax.set_ylabel('frequency')
    mean_val = group['result'].mean()
    ax.text(0.95, 0.95, f'average: {mean_val:.2f}', horizontalalignment='right', verticalalignment='top', transform=ax.transAxes)


plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(4, 3))
sns.boxplot(x='범례', y='result', data=data2)
plt.title('Boxplot of Result by Label')
plt.xlabel('Label')
plt.ylabel('Result')
plt.show()
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# 가정: 'data2'는 pandas Data2Frame이며, 'label'과 'result' 열을 포함하고 있음
# data2 = pd.read_csv('your_data2.csv') # 데이터 파일을 불러오는 경우

# label에 따라 데이터를 분리
group0 = data2[data2['label'] == 0]['result']
group1 = data2[data2['label'] == 1]['result']

# 독립 표본 t-test 수행
t_stat, p_val = ttest_ind(group0, group1)

# 결과 출력
print(f"t-statistic: {t_stat}, p-value: {p_val}")
```
